```python
# Fonction pour l'affichage des textes dans l'output
def print_formated(text, line_length=115):
    """
    Ajoute des retours à la ligne tous les 120 caractères dans une chaîne de caractères pour en faciliter la lecture dans l'output.
    args :
    text (str): La chaîne de caractères à formater.
    line_length (int): La longueur de chaque ligne avant d'ajouter un retour à la ligne, par défaut 120.
    
    returns:
    print(str): La chaîne de caractères formatée avec des retours à la ligne.
    """
    new_text = '\n'.join([text[i:i+line_length] for i in range(0, len(text), line_length)])
    return print(new_text, "\n")

# Nettoyage des te"xtes
def clean_text(text):
    # Supprimer les URLs + les liens spécifique twitter
    text = re.sub(r'www\S+|http\S+', '', text) # Supprimer les URLs 
    text = re.sub(r'pic.twitter.com/\S+', '', text) # Supprimer les liens spécifique twitter
    text = re.sub(r"@\w+|#", "", text) # Supprimer les mentions
    text = re.sub(r'#', '', text) # Supprimer les hashtags
    text = re.sub(r'[^A-Za-z ]+', ' ', text) # Supprimer les caractères non alphabétiquess
    text = re.sub(r'\s+', ' ', text) # Supprimer les espaces multiples
    return text


val_size = 0.1  #  10% des données pour la validation sur le train
test_size=0.2 # 20% des données pour le test sur l'ensemble des données
# focntion de division du dataset
def split_data(X, y, random_state=seed_value):
    """
    Divise le jeu de données en ensembles d'entraînement, de validation et de test.
      
    """
   
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed_value, stratify=y)
    
    X_train_final, X_val, y_train_final, y_val = train_test_split(X_train, y_train, test_size=val_size, random_state=seed_value, stratify=y_train) 

    return X_train_final, X_val, y_train_final, y_val, X_test, y_test


# Fonction de remapping des prédictions
def map_predictions(predictions):
    """
    Remap classe 2 (neutre) à 1 (positif) pour une classification binaire.
    """
    return [0 if pred == 0 else 1 for pred in predictions]
    
```

# Tests modèles BERT pré-entraînés

**Les Transformers**
- Les Transformers sont une architecture de réseau de neurones introduite en 2017 par Vaswani et al. dans l’article “Attention Is All You Need”. Ils reposent principalement sur le mécanisme d’attention, qui permet au modèle de se concentrer sur différentes parties de la séquence d’entrée pour mieux comprendre le contexte global. 
- Contrairement aux réseaux récurrents (RNN/LSTM), les Transformers traitent l’ensemble de la séquence en parallèle, ce qui accélère l’entraînement et améliore la capture des dépendances à longue distance dans le texte.
- Les Transformers sont à la base de nombreux modèles modernes de NLP, dont BERT, GPT, et leurs variantes.

***sources :***
- ***[Attention Is All You Need (arXiv, 2017)](https://arxiv.org/abs/1706.03762)***
- ***[Hugging Face - Transformers](https://huggingface.co/docs/transformers/index)***


**1. BERT : un modèle de langage basé sur les Transformers**
- BERT (Bidirectional Encoder Representations from Transformers) est un modèle de langage développé par Google en 2018, basé sur l’architecture Transformer. Il se distingue par sa capacité à traiter le texte de manière bidirectionnelle, c’est-à-dire en prenant en compte le contexte à gauche et à droite d’un mot simultanément.

**2. Pré-entraînement sur des tâches de compréhension du langage**
- BERT est pré-entraîné sur deux tâches :
    - Masked Language Modeling (MLM) : certains mots de la phrase sont masqués et le modèle doit les prédire à partir du contexte.
    - Next Sentence Prediction (NSP) : le modèle doit prédire si une phrase suit logiquement une autre.
- Ce pré-entraînement permet à BERT de capturer des relations sémantiques et syntaxiques profondes dans le texte.

**3. Fine-tuning pour des tâches spécifiques**
- Après le pré-entraînement, BERT peut être adapté (fine-tuned) à de nombreuses tâches de NLP : classification de texte, analyse de sentiments, questions-réponses, etc. Il suffit d’ajouter une couche de sortie adaptée à la tâche et d’entraîner le modèle sur un jeu de données annoté.

**4. Avantages de BERT**
- Prise en compte du contexte bidirectionnel (contrairement à Word2Vec ou GloVe qui sont unidirectionnels ou context-free).
- Excellentes performances sur de nombreuses tâches de NLP, souvent à l’état de l’art.
- Large disponibilité de modèles pré-entraînés pour différentes langues et domaines.

* BERT est un modèle de langage basé sur l’architecture Transformer, pré-entraîné de façon bidirectionnelle sur de grandes quantités de texte, puis adapté à des tâches spécifiques de NLP. Il a révolutionné le traitement automatique du langage naturel grâce à sa capacité à comprendre le contexte global d’un texte.

**5. Inconvénients de BERT**
- Coût computationnel élevé : L’entraînement et l’inférence avec BERT nécessitent beaucoup de ressources matérielles (GPU/TPU), ce qui peut limiter son utilisation en production ou sur de gros volumes de données.
- Latence et taille du modèle : Les modèles BERT sont volumineux (des centaines de millions de paramètres), ce qui entraîne une latence importante lors de l’inférence et une consommation mémoire élevée.
- Longueur maximale des séquences : BERT ne peut traiter que des séquences de longueur limitée (512 tokens maximum), ce qui peut normalement poser problème pour les textes longs. Dans le cadre de ce projet, cet inconvénient n'aura que peu d'impact. 
- Pas optimal pour les tâches nécessitant une génération de texte : BERT est un modèle d’encodage (encoder-only), il n’est pas conçu pour générer du texte, contrairement à des modèles comme GPT (decoder-only) ou T5 (encoder-decoder).

***sources :***
- ***[BERT: Pre-training of Deep Bidirectional Transformers for Language Understanding](https://arxiv.org/abs/1810.04805)***
- ***[Hugging Face - Transformers - BERT](https://huggingface.co/docs/transformers/model_doc/bert)***

```python
os.environ["MLFLOW_HTTP_REQUEST_TIMEOUT"] = "600"

# Evaluer le modèle BERT avec MLflow
def train_and_evaluate_model_with_bert(model_name, documents, labels, max_length=60, test_size=0.2, batch_size=8, epochs=3):
    """
    Entraîne et évalue un modèle BERT pré-entraîné avec MLflow.
    """
    if mlflow.active_run() is not None:
        mlflow.end_run()
        
   
    print(f"Évaluation du modèle : {model_name}")
    
    print("GPU disponible :", tf.config.list_physical_devices('GPU'))
    print("GPU utilisé :", tf.config.list_logical_devices('GPU'))
    
    
    # Charger la configuration, tokenizer et modèle
    config = AutoConfig.from_pretrained(model_name)
    tokenizer = AutoTokenizer.from_pretrained(model_name)
    try:
        # Charger le modèle TensorFlow
        model = TFAutoModelForSequenceClassification.from_pretrained(model_name, config=config)
    except OSError:
        # Charger le modèle PyTorch si TensorFlow n'est pas disponible
        print(f"Model {model_name} is in PyTorch format, converting to TensorFlow using `from_pt=True`.")
        model = TFAutoModelForSequenceClassification.from_pretrained(model_name, config=config, from_pt=True)

    print(model.summary())
    
    print("Début du préprocessing des données")
    start_time = time.time()
    # Nettoyage des textes
    documents = documents.apply(clean_text) 
    
    print("Exemple après préprocess : \n") 
    print_formated(documents.iloc[0])
    
    # séparer les données en ensemble de train, de validation et de test
    X = documents
    y = labels
    X_train_final, X_val, y_train_final, y_val, X_test, y_test = split_data(X, y, random_state=seed_value)
    
    # Tokenisation des données
    def tokenize_texts(texts):
        encoding = tokenizer(
            texts.tolist(),
            add_special_tokens=True,
            max_length=max_length,
            padding="max_length",
            truncation=True,
            return_tensors="np"
        )
        return encoding["input_ids"], encoding["attention_mask"]
    
    # train_input_ids, train_attention_masks = tokenize_texts(X_train_final)
    val_input_ids, val_attention_masks = tokenize_texts(X_val)
    test_input_ids, test_attention_masks = tokenize_texts(X_test)
    
    preprocessing_time = time.time() - start_time   
    print("Fin du préprocessing des données - Durée:", preprocessing_time, "secondes")

    print("val_input_ids shape:", val_input_ids.shape)
    print("textes tokenisés - ok ")
    print("y_val shape:", y_val.shape)
    print("y_test shape:", y_test.shape)
    
    # Entraîner le modèle avec MLflow with mlflow.start_run(run_name=model_name) as run:
    with mlflow.start_run(run_name=model_name) as run: 
        print("Enregistrement des hyperparamètres - en cours")
        # Enregistrer les hyperparamètres
        mlflow.set_tag("model_name", model_name)
        mlflow.log_param("model_name", model_name)
        mlflow.log_param("max_length", max_length)
        mlflow.log_param("batch_size", batch_size)
        mlflow.log_param("epochs", epochs)
        # mlflow.log_param("test_size", test_size)
        mlflow.log_param("val_size", val_size)
        
        print("Hyperparamètres enregistrés - ok ")

        # Prédictions sur l'ensemble de validation
        val_start_time = time.time()
        outputs_val = model.predict([val_input_ids, val_attention_masks])
        val_predict_time = time.time() - val_start_time
        print("Prédictions sur validation terminées en : ", val_predict_time, "secondes")
        logits_val = outputs_val.logits
        y_val_prob = tf.nn.softmax(logits_val, axis=-1).numpy()
        y_pred_val = np.argmax(y_val_prob, axis=-1)
        
        # Remap des prédictions si le modèle utilise 3 classes
        if config.num_labels == 3:
            print(f"Remapping predictions for binary classification (Model: {model_name})...")
            y_pred_val = map_predictions(y_pred_val)

        # Prédictions sur l'ensemble de test
        test_start_time = time.time()
        outputs_test = model.predict([test_input_ids, test_attention_masks])
        test_predict_time = time.time() - test_start_time
        print("Prédictions sur test terminées en : ", test_predict_time, "secondes")
        logits_test = outputs_test.logits
        y_test_prob = tf.nn.softmax(logits_test, axis=-1).numpy()
        y_pred_test = np.argmax(y_test_prob, axis=-1)
        
         # Remap des prédictions si le modèle utilise 3 classes
        if config.num_labels == 3:
            print(f"Remapping predictions for binary classification (Model: {model_name})...")
            y_pred_test = map_predictions(y_pred_test)
        
        # Calculer les métriques
        accuracy_val = accuracy_score(y_val, y_pred_val)
        auc_val = roc_auc_score(y_val, y_pred_val)
        accuracy_test = accuracy_score(y_test, y_pred_test)
        auc_test = roc_auc_score(y_test, y_pred_test)
        
        print("Calcul des métriques - ok")
        
         # Enregistrer les métriques
        mlflow.log_metric("accuracy_val", accuracy_val)
        mlflow.log_metric("auc_val", auc_val)
        mlflow.log_metric("accuracy_test", accuracy_test)
        mlflow.log_metric("auc_test", auc_test)
        mlflow.log_metric("val_predict_time", val_predict_time)
        mlflow.log_metric("test_predict_time", test_predict_time)
        mlflow.log_metric("preprocessing_time", preprocessing_time)
        
        print('Métriques enregistrées - ok')
        
        # Enregistrer le modèle avec un exemple d'entrée
        input_example = val_input_ids[0]
        
        model.save(f"artifacts/{model_name}")
        if not os.path.exists(f"artifacts/{model_name}"):
            raise FileNotFoundError(f"Le modèle n'a pas été sauvegardé correctement à l'emplacement : artifacts/{model_name}")
        
        # Enregistrer le modèle avec un nom personnalisé et un exemple d'entrée
        model_info = mlflow.tensorflow.log_model(
        artifact_path=f"{model_name}",         # Chemin d'enregistrement dans MLflow
        model=model,         
        input_example=input_example)       
        
        run_id = run.info.run_id
        print(f"Run ID ({model_name}): {run_id}")
        model_uri = model_info.model_uri
        
        def clean_model_name(name):
            return name.replace("/", "-").replace(" ", "_")

        model_name = clean_model_name(model_name)
        
        mlflow.register_model(
            model_uri=f"runs:/{run_id}/model",
            name=f"{model_name}"
        )
        
        print("Modèle enregistré - ok")
        
        # Visualisation des résultats
        print("Affichage des métriques d'évaluation")
        print("Rapport de classification sur la validation: \n", classification_report(y_val, y_pred_val))
        print("Rapport de classification sur le test: \n", classification_report(y_test, y_pred_test))
               
        # Générer et enregistrer la matrice de confusion pour le val
        cm_val = confusion_matrix(y_val, y_pred_val)
        cm_percent_val = cm_val.astype('float') / cm_val.sum(axis=1)[:, np.newaxis] * 100
        cm_display_val = np.array([["{0:.0f}\n({1:.1f}%)".format(value, percent) for value, percent in zip(row, row_percent)] for row, row_percent in zip(cm_val, cm_percent_val)])
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm_val, annot=cm_display_val, fmt='', cmap='Greens', xticklabels=[target_labels[0], target_labels[1]], yticklabels=[target_labels[0], target_labels[1]])
        plt.xlabel('Prédit')
        plt.ylabel('Réel')
        plt.title(f'Matrice de Confusion {model_name} (val)')
        plt.savefig(f"./Graph_eval/Conf_Mat_val/confusion_matrix_{model_name}_val.png")
        mlflow.log_artifact(f"./Graph_eval/Conf_Mat_val/confusion_matrix_{model_name}_val.png")

        # Générer et enregistrer la matrice de confusion pour le test
        cm_test = confusion_matrix(y_test, y_pred_test)
        cm_percent_test = cm_test.astype('float') / cm_test.sum(axis=1)[:, np.newaxis] * 100
        cm_display_test = np.array([["{0:.0f}\n({1:.1f}%)".format(value, percent) for value, percent in zip(row, row_percent)] for row, row_percent in zip(cm_test, cm_percent_test)])
        plt.figure(figsize=(8, 6))
        sns.heatmap(cm_test, annot=cm_display_test, fmt='', cmap='YlOrBr', xticklabels=[target_labels[0], target_labels[1]], yticklabels=[target_labels[0], target_labels[1]])
        plt.xlabel('Prédit')
        plt.ylabel('Réel')
        plt.title(f'Matrice de Confusion {model_name} (Test)')
        plt.savefig(f"./Graph_eval/Conf_Mat_Test/confusion_matrix_{model_name}_test.png")
        mlflow.log_artifact(f"./Graph_eval/Conf_Mat_Test/confusion_matrix_{model_name}_test.png")

        # Générer et enregistrer la courbe ROC
        fpr, tpr, _ = roc_curve(y_test, y_pred_test)
        plt.figure()
        plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_test)
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {model_name}')
        plt.legend(loc="lower right")
        plt.savefig(f"./Graph_eval/Roc_curve/roc_curve_{model_name}.png")
        mlflow.log_artifact(f"./Graph_eval/Roc_curve/roc_curve_{model_name}.png")
        
         # Générer et enregistrer la courbe ROC
        fpr, tpr, _ = roc_curve(y_val, y_pred_val)
        plt.figure()
        plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_val)
        plt.plot([0, 1], [0, 1], 'k--')
        plt.xlim([0.0, 1.0])
        plt.ylim([0.0, 1.05])
        plt.xlabel('False Positive Rate')
        plt.ylabel('True Positive Rate')
        plt.title(f'ROC Curve for {model_name}')
        plt.legend(loc="lower right")
        plt.savefig(f"./Graph_eval/Roc_curve/roc_curve_{model_name}.png")
        mlflow.log_artifact(f"./Graph_eval/Roc_curve/roc_curve_{model_name}.png")

            
        # Afficher le run_id
        run_id = run.info.run_id
        print(f"Run ID ({model_name}): {run_id}")

```

```python
if mlflow.active_run() is not None:
        mlflow.end_run()
        
# model_name = f"BERT_{model_name}_reducted_dataset"        
print(f"Évaluation du modèle : {model_name}")

print("GPU disponible :", tf.config.list_physical_devices('GPU'))
print("GPU utilisé :", tf.config.list_logical_devices('GPU'))

# Charger le tokenizer et le modèle
model_name = 'finiteautomata/bertweet-base-sentiment-analysis'
tokenizer = AutoTokenizer.from_pretrained(model_name)

print("Début du préprocessing des données")
# Nettoyage des te"xtes
def clean_text(text):
    # Supprimer les URLs + les liens spécifique twitter
    text = re.sub(r'www\S+|http\S+', '', text) # Supprimer les URLs 
    text = re.sub(r'pic.twitter.com/\S+', '', text) # Supprimer les liens spécifique twitter
    text = re.sub(r"@\w+|#", "", text) # Supprimer les mentions
    text = re.sub(r'#', '', text) # Supprimer les hashtags
    text = re.sub(r'[^A-Za-z ]+', ' ', text) # Supprimer les caractères non alphabétiquess
    text = re.sub(r'\s+', ' ', text) # Supprimer les espaces multiples
    text = emoji.demojize(text) # Supprimer les emojis
    return text

start_time = time.time()
# Nettoyage des textes
documents = documents.apply(clean_text) 

print("Exemple après préprocess : \n") 
print_formated(documents.iloc[0])

# séparer les données en ensemble de train, de validation et de test
X = documents
y = labels
X_train, X_val, y_train, y_val, X_test, y_test = split_data(X, y, random_state=seed_value)   

# Tokenisation des données
def tokenize_texts(texts, max_length=80):
    return tokenizer(
        texts.tolist(),
        add_special_tokens=True,
        max_length=max_length,
        padding='max_length',
        truncation=True,
        return_tensors="tf"
    )

train_encodings = tokenize_texts(X_train)
val_encodings = tokenize_texts(X_val)
test_encodings = tokenize_texts(X_test)

# Préparer les datasets TensorFlow
train_dataset = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train.values
)).shuffle(len(X_train)).batch(16)

# Pour la prédiction sur le train (pas de shuffle !)
train_dataset_noshuffle = tf.data.Dataset.from_tensor_slices((
    dict(train_encodings),
    y_train.values
)).batch(16)

val_dataset = tf.data.Dataset.from_tensor_slices((
    dict(val_encodings),
    y_val.values
)).batch(16)

test_dataset = tf.data.Dataset.from_tensor_slices((
    dict(test_encodings),
    y_test.values
)).batch(16)

preprocessing_time = time.time() - start_time   
print("Fin du préprocessing des données - Durée:", preprocessing_time, "secondes")

# Charger la configuration et le modèle pré-entraîné sans la couche de classification
config = AutoConfig.from_pretrained(model_name, output_hidden_states=False)
base_model = TFAutoModel.from_pretrained(model_name, config=config)

# Dégèle de la dernieère couche du modèle pour accélérer l'entraînement
for layer in base_model.layers[:-1]:
    layer.trainable = False

# Créer un nouveau modèle Keras
input_ids = Input(shape=(80,), dtype=tf.int32, name="input_ids")
attention_mask = Input(shape=(80,), dtype=tf.int32, name="attention_mask")

# Utiliser Lambda pour appeler le modèle pré-entraîné avec output_shape spécifié
sequence_output = Lambda(
    lambda inputs: base_model(input_ids=inputs[0], attention_mask=inputs[1])[0],
    output_shape=(80, 768)
)([input_ids, attention_mask])

cls_token = Lambda(lambda x: x[:, 0, :], output_shape=(768,))(sequence_output)  # Extraire le token [CLS]

# Ajouter une couche dense pour la classification binaire
output = Dense(1, activation="sigmoid")(cls_token)

# Construire le modèle final
model = Model(inputs=[input_ids, attention_mask], outputs=output)
print(model.summary())

# Optimiseur et perte
learning_rate = 1e-5
optimizer = Adam()
loss = tf.keras.losses.BinaryCrossentropy(from_logits=False)
metric = tf.keras.metrics.BinaryAccuracy('accuracy')

# Compiler le modèle
model.compile(optimizer=optimizer, loss=loss, metrics=[metric])

# Entraîner le modèle avec MLflow with mlflow.start_run(run_name=model_name) as run:
experiment = mlflow.set_experiment("BERT_finetuning")
with mlflow.start_run(run_name="BERT_finetuning") as run:
    
    print("Enregistrement des hyperparamètres - en cours")
    batch_size=16
    epochs=3
    # Enregistrer les hyperparamètres
    mlflow.set_tag("model_name", model_name)
    mlflow.log_param("model_name", model_name)
    mlflow.log_param("max_length", max_length)
    mlflow.log_param("batch_size", batch_size)
    mlflow.log_param("epochs", epochs)
    # mlflow.log_param("test_size", test_size)
    mlflow.log_param("val_size", val_size)
    
    print("Hyperparamètres enregistrés - ok ")
    
    print("Entraînement du modèle...")
    start_time = time.time()
    # Entraîner le modèle
    history = model.fit(
        train_dataset,
        validation_data=val_dataset,
        epochs=epochs,
        batch_size=batch_size
    )
    fit_time = time.time() - start_time
    print("Entraînement terminé en : ", fit_time, "secondes")
    
    # Prédictions sur l'ensemnle de validation
    start_time = time.time()
    logits_train = model.predict(train_dataset_noshuffle)
    train_predict_time = time.time() - start_time
    print("Prédictions sur train_dataset terminées en : ", train_predict_time, "secondes")
    # Convertir les probabilités en classes binaires
    y_train_prob = logits_train.flatten()  # Pour roc_auc_score, garder les probabilités
    y_pred_train = (y_train_prob > 0.5).astype(int) 
    
    # Prédictions sur l'ensemnle de validation
    start_time = time.time()
    logits_val = model.predict(val_dataset)
    val_predict_time = time.time() - start_time
    print("Prédictions sur val_dataset terminées en : ", val_predict_time, "secondes")
    # Convertir les probabilités en classes binaires
    y_val_prob = logits_val.flatten()  # Pour roc_auc_score, garder les probabilités
    y_pred_val = (y_val_prob > 0.5).astype(int) 
    
    # Prédictions sur l'ensemnle de test
    start_time = time.time()
    logits_test = model.predict(test_dataset)
    test_predict_time = time.time() - start_time
    print("Prédictions sur test_dataset terminées en : ", test_predict_time, "secondes")
    # Convertir les probabilités en classes binaires
    y_test_prob = logits_test.flatten()  # Pour roc_auc_score, garder les probabilités
    y_pred_test = (y_test_prob > 0.5).astype(int) 
    
    # Calculer les métriques
    accuracy_train = accuracy_score(y_train, y_pred_train)
    auc_train = roc_auc_score(y_train, y_pred_train)
    accuracy_val = accuracy_score(y_val, y_pred_val)
    auc_val = roc_auc_score(y_val, y_pred_val)
    accuracy_test = accuracy_score(y_test, y_pred_test)
    auc_test = roc_auc_score(y_test, y_pred_test)
    
    print("Calcul des métriques - ok")
    
    # Enregistrer les métriques
    mlflow.log_metric("accuracy_train", accuracy_train)
    mlflow.log_metric("auc_train", auc_train)
    mlflow.log_metric("accuracy_val", accuracy_val)
    mlflow.log_metric("auc_val", auc_val)
    mlflow.log_metric("accuracy_test", accuracy_test)
    mlflow.log_metric("auc_test", auc_test)
    mlflow.log_metric("fit_time", fit_time)
    mlflow.log_metric("train_predict_time", train_predict_time)
    mlflow.log_metric("val_predict_time", val_predict_time)
    mlflow.log_metric("test_predict_time", test_predict_time)
    mlflow.log_metric("preprocessing_time", preprocessing_time)
    
    print('Métriques enregistrées - ok')
    
    model.save(f"artifacts/{model_name}_BERT_finetuning.keras")
    if not os.path.exists(f"artifacts/{model_name}_BERT_finetuning.keras"):
        raise FileNotFoundError(f"Le modèle n'a pas été sauvegardé correctement à l'emplacement : artifacts/{model_name}_BERT_finetuning.keras")
    
    # Enregistrer le modèle avec un nom personnalisé 
    input_example = documents.iloc[0]
    model_info = mlflow.tensorflow.log_model(
    artifact_path=f"{model_name}",         # Chemin d'enregistrement dans MLflow
    model=model,        
    input_example=input_example
    )         
    
    run_id = run.info.run_id
    print(f"Run ID ({model_name}): {run_id}")
    model_uri = model_info.model_uri
    
    def clean_model_name(name):
        return name.replace("/", "-").replace(" ", "_")

    model_name = clean_model_name(model_name)
    
    mlflow.register_model(
        model_uri=f"runs:/{run_id}/model",
        name=f"{model_name}"
    )
    
    print("Modèle enregistré - ok")
        
    # Visualisation des résultats
    print("Affichage des métriques d'évaluation")
    print("Rapport de classification sur le train: \n", classification_report(y_train, y_pred_train))
    print("Rapport de classification sur la validation: \n", classification_report(y_val, y_pred_val))
    print("Rapport de classification sur le test: \n", classification_report(y_test, y_pred_test))
    
    # Générer et enregistrer la matrice de confusion pour le train
    cm_train = confusion_matrix(y_train, y_pred_train)
    cm_percent_train = cm_train.astype('float') / cm_train.sum(axis=1)[:, np.newaxis] * 100
    cm_display_train = np.array([["{0:.0f}\n({1:.1f}%)".format(trainue, percent) for trainue, percent in zip(row, row_percent)] for row, row_percent in zip(cm_train, cm_percent_train)])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_train, annot=cm_display_train, fmt='', cmap='Blues', xticklabels=[target_labels[0], target_labels[1]], yticklabels=[target_labels[0], target_labels[1]])
    plt.xlabel('Prédit')
    plt.ylabel('Réel')
    plt.title(f'Matrice de Confusion {model_name}_BERT_finetuning (train)')
    plt.savefig(f"./Graph_eval/Conf_Mat_train/confusion_matrix_{model_name}_BERT_finetuning_train.png")
    mlflow.log_artifact(f"./Graph_eval/Conf_Mat_train/confusion_matrix_{model_name}_BERT_finetuning_train.png")
    
    # Générer et enregistrer la matrice de confusion pour le val
    cm_val = confusion_matrix(y_val, y_pred_val)
    cm_percent_val = cm_val.astype('float') / cm_val.sum(axis=1)[:, np.newaxis] * 100
    cm_display_val = np.array([["{0:.0f}\n({1:.1f}%)".format(value, percent) for value, percent in zip(row, row_percent)] for row, row_percent in zip(cm_val, cm_percent_val)])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_val, annot=cm_display_val, fmt='', cmap='Greens', xticklabels=[target_labels[0], target_labels[1]], yticklabels=[target_labels[0], target_labels[1]])
    plt.xlabel('Prédit')
    plt.ylabel('Réel')
    plt.title(f'Matrice de Confusion {model_name}_BERT_finetuning (val)')
    plt.savefig(f"./Graph_eval/Conf_Mat_val/confusion_matrix_{model_name}_BERT_finetuning_val.png")
    mlflow.log_artifact(f"./Graph_eval/Conf_Mat_val/confusion_matrix_{model_name}_BERT_finetuning_val.png")

    # Générer et enregistrer la matrice de confusion pour le test
    cm_test = confusion_matrix(y_test, y_pred_test)
    cm_percent_test = cm_test.astype('float') / cm_test.sum(axis=1)[:, np.newaxis] * 100
    cm_display_test = np.array([["{0:.0f}\n({1:.1f}%)".format(value, percent) for value, percent in zip(row, row_percent)] for row, row_percent in zip(cm_test, cm_percent_test)])
    plt.figure(figsize=(8, 6))
    sns.heatmap(cm_test, annot=cm_display_test, fmt='', cmap='YlOrBr', xticklabels=[target_labels[0], target_labels[1]], yticklabels=[target_labels[0], target_labels[1]])
    plt.xlabel('Prédit')
    plt.ylabel('Réel')
    plt.title(f'Matrice de Confusion {model_name}_BERT_finetuning (Test)')
    plt.savefig(f"./Graph_eval/Conf_Mat_Test/confusion_matrix_{model_name}_BERT_finetuning_test.png")
    mlflow.log_artifact(f"./Graph_eval/Conf_Mat_Test/confusion_matrix_{model_name}_BERT_finetuning_test.png")

    # Générer et enregistrer la courbe ROC
    fpr, tpr, _ = roc_curve(y_val, y_pred_val)
    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_val)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Validation - ROC Curve for {model_name}_BERT_finetuning')
    plt.legend(loc="lower right")
    plt.savefig(f"./Graph_eval/Roc_curve/roc_curve_{model_name}_BERT_finetuning.png")
    mlflow.log_artifact(f"./Graph_eval/Roc_curve/roc_curve_{model_name}_BERT_finetuning.png")
    
    # Générer et enregistrer la courbe ROC
    fpr, tpr, _ = roc_curve(y_test, y_pred_test)
    plt.figure()
    plt.plot(fpr, tpr, label='ROC curve (area = %0.2f)' % auc_test)
    plt.plot([0, 1], [0, 1], 'k--')
    plt.xlim([0.0, 1.0])
    plt.ylim([0.0, 1.05])
    plt.xlabel('False Positive Rate')
    plt.ylabel('True Positive Rate')
    plt.title(f'Test - ROC Curve for {model_name}_BERT_finetuning')
    plt.legend(loc="lower right")
    plt.savefig(f"./Graph_eval/Roc_curve/roc_curve_{model_name}_BERT_finetuning.png")
    mlflow.log_artifact(f"./Graph_eval/Roc_curve/roc_curve_{model_name}_BERT_finetuning.png")
    
   

```
